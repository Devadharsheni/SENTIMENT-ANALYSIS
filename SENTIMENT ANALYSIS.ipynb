```python
import nltk
from nltk.corpus import stopwords
import re
import string
from nltk.tokenize import TweetTokenizer

nltk.download('stopwords',quiet=True)
english_stopwords=stopwords.words('english')

def clean_tweet(tweet):
    # Handle the case where tweet is not a string
    if not isinstance(tweet, str):
        return ""  # Or any appropriate handling for non-string tweets

    tweet=re.sub(r"https\S+|www\S+|@\w+|#\w+","",tweet) # Remove URLs, mentions, and hashtags
    tweet=re.sub(r"[^\w\s]","",tweet) # Remove punctuation
    tokenizer=TweetTokenizer(preserve_case=False,reduce_len=True,strip_handles=True)
    tokens=tokenizer.tokenize(tweet)
    tokens=[token for token in tokens if token not in english_stopwords] # Remove stopwords
    tokens=[token.translate(str.maketrans('','',string.punctuation))for token in tokens] # Remove punctuation again for any missed characters
    tokens=[token.lower() for token in tokens] # Convert to lowercase
    cleaned_tweet=' '.join(tokens)
    return cleaned_tweet
```
